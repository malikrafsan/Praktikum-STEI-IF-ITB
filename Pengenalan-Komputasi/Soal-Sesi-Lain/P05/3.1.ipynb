```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('crypto.csv')
```

```python
df.head()
```

```python
df.loc[df['name'] == 'BTC'].sort_values('date').plot(kind='line', x='date', y='high')


df.loc[df['name'] == 'BTC'].sort_values('date').plot.line(x='date', y='high')
```

```python
for i in ['BTC', 'ETH', 'XRP', 'USDT', 'XMR']:
    df.loc[df['name'] == i].sort_values('date').plot.line(x='date', y='high', title=i, figsize=(10,5))
```

```python
df.groupby('name').count()['market'].plot(kind='pie')
```
