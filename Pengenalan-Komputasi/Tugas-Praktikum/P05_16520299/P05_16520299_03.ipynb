## Membaca data

```python
# Mengimport library pandas dan matplotlib.pyplot
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("crypto.csv")
```

## Line Chart

```python
# Jawaban 1
# Line chart untuk harga ETH setiap harinya

df.loc[df["name"] == "ETH"].plot(kind="line", x="date", y="close", title="Line Chart Harga ETH Setiap Harinya")
```

## Scatter Plot

```python
# Jawaban 2
# Scatter plot dengan nilai high sebagai sumbu x dan nilai low sebagai sumbu y untuk koin USDT

df.loc[df["name"] == "USDT"].plot(kind="scatter", x="high", y="low", title="Scatter Plot Nilai High Vs Nilai Low")
```

## Pie Chart

```python
# Jawaban 3
# Pie chart banyaknya data masing-masing koin

df["name"].value_counts().plot(kind="pie", title="Pie Chart Banyaknya Data Masing-Masing Koin")
```

```python
# Jawaban 4
# Pie chart nilai kapitalisasi pasar semua mata uang pada tanggal 2 Maret 2015

df4 = df.loc[df["date"] == "2015-03-02"].groupby("name").sum()
df4.plot(kind="pie", y="market", title="Pie Chart Nilai Kapitalisasi Pasar Semua Mata Uang pada Tanggal 2 Maret 2015")
```

## Line Chart

```python
# Jawaban 5
# Line chart untuk membandingkan nilai close BTC, ETH, XRP, USDT, dan XMR

df.groupby(["date", "name"])["close"].sum().unstack().plot(kind="line", 
                                                           title="Line Chart Nilai Close BTC, ETH, XRP, USDT, dan XMR")
```
