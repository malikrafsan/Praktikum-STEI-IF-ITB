## Membaca Data

```python
# Mengimport library pandas
import pandas as pd

df = pd.read_csv("crypto.csv")
```

## Mengakses Data

```python
# Jawaban 1
# Data untuk koin XMR

df.loc[df["name"] == "XMR"]
```

```python
# Jawaban 2
# Data yang ada pada tanggal 2 Maret 2015

df.loc[df["date"] == "2015-03-02"]
```

```python
# Jawaban 3
# 5 data dengan nilai low terendah

df.sort_values("low")[:5]
```

```python
# Jawaban 4
# Nilai high tertinggi

print("Nilai high tertinggi adalah " + str(df["high"].max()))
```

```python
# Jawaban 5
# Tanggal ketika nilai high tertinggi XMR terjadi

XMRmax = df.loc[df["name"] == "XMR"]["high"].max()
print("Tanggal ketika nilai high tertinggi XMR terjadi " + 
      str(df.loc[(df["name"] == "XMR") & (df["high"] == XMRmax)]["date"].to_string(index=False)))
```
